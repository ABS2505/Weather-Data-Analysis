```python
df['datetime'] = pd.to_datetime(df['datetime'])
df.set_index('datetime', inplace=True)
```

```python
daily_temp = df['New York'].resample('D').mean()
```

```python
df['avg_temp'] = df.mean(axis=1)
df.head()
```

```python
daily_temp = df['avg_temp'].resample('D').mean()
daily_temp.head()
```

```python
monthly_temp = df['avg_temp'].resample('ME').mean()
monthly_temp.head()
```

```python
max_temp = df['avg_temp'].max()
min_temp = df['avg_temp'].min()

print("Maximum Temperature:", max_temp)
print("Minimum Temperature:", min_temp)
```

```python
plt.figure(figsize=(12,6))
daily_temp.plot()
plt.title("Daily Average Temperature Trend")
plt.xlabel("Date")
plt.ylabel("Temperature")
plt.show()
```

```python
plt.figure(figsize=(12,6))
monthly_temp.plot(color='orange')
plt.title("Monthly Average Temperature Trend")
plt.xlabel("Month")
plt.ylabel("Temperature")
plt.show()
```

```python
df['month'] = df.index.month
df['year'] = df.index.year
```

```python
pivot = df.pivot_table(values='avg_temp',
                       index='month',
                       columns='year',
                       aggfunc='mean')
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(pivot, cmap='coolwarm')
plt.title("Seasonal Temperature Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['avg_temp'], bins=30)
plt.title("Temperature Distribution")
plt.xlabel("Temperature")
plt.ylabel("Frequency")
plt.show()
```
